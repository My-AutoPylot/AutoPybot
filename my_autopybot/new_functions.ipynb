```python
#PDF Image to Text
def pdf_image_to_text(pdf_file_path : str) -> str:
    """
    Description: Converts PDF Image to Text

    Parameters:
    pdf_file (str): File path of the PDF file
    show_output (bool): If True, shows the output image
    
    Returns:
    None

    Example:
    >>> pdf_image_to_text("test.pdf", True)
    None
    
    """
    # Import section
    import pytesseract
    from PIL import Image
    import pdf2image
    import os

    # Code section
    text = ""

    pages = pdf2image.convert_from_path(pdf_file_path, 500)
    for page in pages:
        page.save('out.jpg', 'JPEG')
        text = pytesseract.image_to_string(Image.open('out.jpg'))
        os.remove('out.jpg')
        
    return text

print(pdf_image_to_text(r"C:/Users/mrmay/Downloads/non-text-searchable.pdf"))
```

```python
#file copy from source to destination, handle permission error
def copy_file(source_file_path : str, destination_file_path : str) -> None:
    """
    Description: Copies file from source to destination

    Parameters:
    source_file_path (str): File path of the source file
    destination_file_path (str): File path of the destination file
    
    Returns:
    None

    Example:
    >>> copy_file("source.txt", "destination.txt")
    None
    
    """
    # Import section
    import shutil
    import os
    
    # Code section
    try:
        shutil.copy2(source_file_path, destination_file_path)
    except PermissionError:
        try:
            os.chmod(destination_file_path, 0o777)
            shutil.copy2(source_file_path, destination_file_path)
        except PermissionError:
            print("Permission Error: Please check the permissions for the destination folder")
            #execute in elevated mode        

copy_file(r"D:\F1",r"D:\F2")
```

```python
#file move from source to destination, handle permission error
def move_file(source_file_path : str, destination_file_path : str) -> None:
    """
    Description: Moves file from source to destination

    Parameters:
    source_file_path (str): File path of the source file
    destination_file_path (str): File path of the destination file
    
    Returns:
    None

    Example:
    >>> move_file("source.txt", "destination.txt")
    None
    
    """
    # Import section
    import shutil
    import os

    # Code section
    try:
        shutil.move(source_file_path, destination_file_path)
    except PermissionError:
        os.chmod(destination_file_path, 0o777)
        shutil.move(source_file_path, destination_file_path)


move_file(r"D:\F1",r"D:\F2")
```

```python
#close all open applications
def close_all_open_applications() -> None:
    """
    Description: Closes all open applications

    Parameters:
    None
    
    Returns:
    None

    Example:
    >>> close_all_open_applications()
    None
    
    """
    # Import section
    import os
    import psutil
    import subprocess

    # Code section
    for process in psutil.process_iter():
        try:
            if str(process.name()).lower() == "excel.exe":
                process.kill()
            elif str(process.name()).lower() == "winword.exe":
                process.kill()
            elif str(process.name()).lower() == "powerpnt.exe":
                process.kill()
            elif str(process.name()).lower() == "outlook.exe":
                process.kill()
            elif str(process.name()).lower() == "chrome.exe":
                process.kill()
            elif str(process.name()).lower() == "firefox.exe":
                process.kill()
            elif str(process.name()).lower() == "msedge.exe":
                process.kill()
            elif str(process.name()).lower() == "iexplore.exe":
                process.kill()
            elif str(process.name()).lower() == "safari.exe":
                process.kill()
            elif str(process.name()).lower() == "opera.exe":
                process.kill()
            elif str(process.name()).lower() == "acrobat.exe":
                process.kill()
            elif str(process.name()).lower() == "acrotray.exe":
                process.kill()
            elif str(process.name()).lower() == "acrord32.exe":
                process.kill()
            elif str(process.name()).lower() == "notepad.exe":
                process.kill()
            elif str(process.name()).lower() == "notepad++.exe":
                process.kill()
            elif str(process.name()).lower() == "wordpad.exe":
                process.kill()
            
        except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess):
            pass

close_all_open_applications()
```

```python
#reverse a string
def reverse_string(string : str) -> str:
    """
    Description: Reverses a string

    Parameters:
    string (str): String
    
    Returns:
    str: Reversed string

    Example:
    >>> reverse_string("string")
    "gnirts"
    
    """
    # Import section
    import re

    # Code section
    return re.sub(r'(?<=\w)(?=\W)|(?<=\W)(?=\w)', '', string[::-1])

reverse_string("string") 
```

```python
#string replace
def string_replace(string : str, old : str, new : str, ignore_case : bool = True, occurance : int = 0) -> str:
    """
    Description: Replaces a string

    Parameters:
    string (str): String
    old (str): Old string
    new (str): New string
    
    Returns:
    str: Replaced string

    Example:
    >>> string_replace("string", "str", "new")
    "newing"
    
    """
    # Import section
    import re

    # Code section
    if ignore_case:
        return re.sub(old, new, string, flags=re.IGNORECASE, count=occurance)
    else:
        return re.sub(old, new, string, count=occurance)
    
# string_replace("string STR string", "str", "new", occurance=0, ignore_case=False)
# string_replace("string STR string", "str", "new", occurance=0)
```

```python
#crop text
def crop_text(text : str, length : int) -> str:
    """
    Description: Crops text

    Parameters:
    text (str): Text
    length (int): Length of text
    
    Returns:
    str: Cropped text

    Example:
    >>> crop_text("string", 3)
    "str"
    
    """
    # Import section
    import re

    # Code section
    return re.sub(r'(?<=\w)(?=\W)|(?<=\W)(?=\w)', '', text[:length])

crop_text("string", -1)    
```

```python
#split text
def split_text(text : str, length : int) -> list:
    """
    Description: Splits text

    Parameters:
    text (str): Text
    length (int): Length of text
    
    Returns:
    list: List of splitted text

    Example:
    >>> split_text("string", 3)
    ["str", "ing"]
    
    """
    # Import section
    import re

    # Code section
    return re.findall(r'.{1,'+str(length)+'}', text)

split_text("my string", 2)
```

```python
#parse text
def parse_text(text : str, start : str, end : str) -> str:
    """
    Description: Parses text

    Parameters:
    text (str): Text
    start (str): Start string
    end (str): End string
    
    Returns:
    str: Parsed text

    Example:
    >>> parse_text("my string", "my", "ing")
    "my str"
    
    """
    # Import section
    import re

    # Code section
    return re.search(start+'(.*)'+end, text).group(1)

parse_text("my string", "my", "ing")    
```

```python
def get_exif_data_from_image(image : str) -> dict: #from exif import Image
    """
    Description: Gets EXIF data from image

    Parameters:
    image (str): Image

    Returns:
    dict: EXIF data

    Example:
    >>> get_exif_data_from_image("image.jpg")

    """
    # Import section
    from exif import Image

    # Code section
    #check if image has EXIF
    if Image(image).has_exif:
        with open(image, 'rb') as image_file:
            my_image = Image(image_file)
            list_all =  my_image.list_all()

            #get all EXIF data
            exif_data = {}
            for i in list_all:
                exif_data[i] = my_image[i]

            return exif_data

# print(get_exif_data_from_image(r"C:\Users\mrmay\OneDrive\Desktop\a5652ea8992f4e658fd17ba07583f5a5.jpg"))
print(get_exif_data_from_image(r"C:\Users\mrmay\OneDrive\Desktop\img\WhatsApp Image 2022-11-05 at 2.01.14 PM - Copy.jpeg"))
```

```python
#translate image x pixels right/left and y pixels up/down using imutils
def translate_image(image_path : Union[str, WindowsPath], x : int, y : int) -> str:
    """
    Description: Translates image x pixels right and y pixels up using imutils

    Parameters:
    image_path (Union[str, WindowsPath]): Image path
    x (int): X pixels right (+) or left (-)
    y (int): Y pixels up (-) or down (+)

    Returns:
    str: Translated image

    Example:
    >>> translate_image("image.jpg", 100, 100)

    """
    # Import section
    import cv2
    import imutils

    # Code section
    image = cv2.imread(image_path)
    image = imutils.translate(image, x, y)

    #get folder from image_path
    folder_path = os.path.dirname(image_path)

    translated = imutils.translate(image, x, y)
    cv2.imwrite(os.path.join(folder_path, "translated.jpg"), translated)
        
# translate_image(r"C:\Users\mrmay\OneDrive\Desktop\Modi\M1.png", -50, -50)
```

```python
#rotate image by angle using imutils

def rotate_image(image_path : Union[str, WindowsPath], angle : int) -> str:
    """
    Description: Rotates image by angle using imutils

    Parameters:
    image_path (Union[str, WindowsPath]): Image path
    angle (int): Angle

    Returns:
    str: Rotated image

    Example:
    >>> rotate_image("image.jpg", 90)

    """
    # Import section
    import cv2
    import imutils

    # Code section
    image = cv2.imread(image_path)
    rotated = imutils.rotate(image, angle)

    #get folder from image_path
    folder_path = os.path.dirname(image_path)

    cv2.imwrite(os.path.join(folder_path, "rotated.jpg"), rotated)

rotate_image(r"C:\Users\mrmay\OneDrive\Desktop\Modi\M1.png", 360)    
```

```python
#Skeletonization of image using imutils
def skeletonize_image(image_path : Union[str, WindowsPath]) -> str:
    """
    Description: Skeletonizes image using imutils

    Parameters:
    image_path (Union[str, WindowsPath]): Image path

    Returns:
    str: Skeletonized image

    Example:
    >>> skeletonize_image("image.jpg")

    """
    # Import section
    import cv2
    import imutils

    # Code section
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    skeletonized = imutils.skeletonize(gray, size=(3, 3))

    #get folder from image_path
    folder_path = os.path.dirname(image_path)

    cv2.imwrite(os.path.join(folder_path, "skeletonized.jpg"), skeletonized)

skeletonize_image(r"C:\Users\mrmay\OneDrive\Desktop\Modi\M1.png")    
```

```python
#Automatic Canny Edge Detection using imutils
def automatic_canny_edge_detection(image_path : Union[str, WindowsPath]) -> str:
    """
    Description: Automatic Canny Edge Detection using imutils

    Parameters:
    image_path (Union[str, WindowsPath]): Image path

    Returns:
    str: Canny Edge Detected image

    Example:
    >>> automatic_canny_edge_detection("image.jpg")

    """
    # Import section
    import cv2
    import imutils

    # Code section
    image = cv2.imread(image_path)
    edged = imutils.auto_canny(image)

    #get folder from image_path
    folder_path = os.path.dirname(image_path)

    cv2.imwrite(os.path.join(folder_path, "edged.jpg"), edged)

automatic_canny_edge_detection(r"C:\Users\mrmay\OneDrive\Desktop\Modi\M1.png")    
```

```python
# doc parser using pdfminer
def doc_parser(file_path : Union[str, WindowsPath]) -> str:
    """
    Description: Parses doc file using pdfminer

    Parameters:
    file_path (Union[str, WindowsPath]): File path

    Returns:
    str: Parsed doc file

    Example:
    >>> doc_parser("file.doc")

    """
    # Import section
    import pdfminer
    import pdfminer.high_level

    # Code section
    text = pdfminer.high_level.extract_text(file_path)

    #get folder from file_path
    folder_path = os.path.dirname(file_path)

    with open(os.path.join(folder_path, "parsed_doc.txt"), "w") as f:
        f.write(text)

doc_parser(r"C:\Users\mrmay\Downloads\OoPdfFormExample.pdf")





```
